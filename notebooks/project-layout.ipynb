This project is a fake news binary classifier , which will classify fake vs real news 

Phase 1 : 
-> collect data from kaggle
-> Acquire labeled articles : headlines , text , author
-> label cleaning : remove duplicates , remove non english , filter very short text
-> save under data/raw


Phase 2:
-> Exploratory data analysis
-> class balance check 
-> text length distribution
-> source distribution (if any)
-> vocab size , top words(bigram , trigram)
-> check label leakage 
-> class counts , word cloud , length histograms
-> check for identical headlines in both classes

Phase 3:
-> preprocessing and feature engineering
-> Text cleaning 
-> Tokenization
-> Build vocab (limit voc_size)
-> pad sequence
-> load pre trained embedding or build from scratch

Phase 4:
-> Model design & baseline 
-> simple or bi lstm + dropout + dense classification head

Phase 5:
-> Training , validation and evaluation
-> Train with class weights
-> Use validation set for early stopping and model selection
-> Traxk various metrics
-> save best checkpoints

Phase 6 :
-> explainibility and QA 
-> testing
-> bias checks : evaluate on different sources/dates


Phase 7 :
-> Inference pipelines and API
-> create pipeline
-> buuild fastapi endpoints
-> add input validation

Phase 8 :
-> Contanerization and deployement
-> Docker file 
-> Local testing
-> kubernets (optional)


Phase 9 :
-> CI/CD 
-> Model Monitioring and retraining loop
-> Github actions
-> Logging and alerts
-> Feedback loop


Phase 10 :
-> Maintainance and improvements
-> Fine tuning
